```
import cv2
import numpy as np
```

```
cap=cv2.VideoCapture(0)
```

```
def sketch(image):

#convert image to grey image
  grey_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)


#Clean image using Gaussian Blur
  grey_image_blur=cv2.GaussianBlur(grey_image,(5,5),0)


#Extract Edges
  canny_edge=cv2.Canny(grey_image_blur,10,70)


#Invert Binarize the Image
  ret,sketch=cv2.threshold(canny_edge,70,255,cv2.THRESH_BINARY_INV)


  return sketch

# If want to covert a sample image to the sketching image 
image_path="/content/IMG_20191203_170939.jpg"

image=cv2.imread(image_path)

cv2.imwrite("Live Sketch.jpg", sketch(image))

#If want to convert the video camera video to sketch feeds
# while True:
#   ret, image=cap.read()
#   cv2.imshow("Live Sketch", sketch(image))


#   if cv2.waitKey(1) == 13:
#       break
    
# cap.release()

# cv2.destroyAllWindows()
    
```
